```python
# Function to convert from Object to date
def conv_obj_date(df:pd.DataFrame,att:object,format=None,error="coerce"):
    df[att]=pd.to_datetime(df[att],format=format,errors=error)
    return df

#  Function to convert from object to int
def conv_obj_int(df:pd.DataFrame,att:object,error="coerce"):
    df[att]=pd.to_numeric(df[att],errors=error).astype(int)
    return df

#  Function to convert from object to float
def conv_obj_float(df:pd.DataFrame,att:object,error="coerce"):
    df[att]=pd.to_numeric(df[att],errors=error).astype(float)
    return df
```
